```python
# plotting testing capacity
df['date'] = pd.to_datetime(df['date'], dayfirst = True)
df.sort_values(by = 'date', ascending = True, inplace = True)


width = 6
height = 5
plt.figure(figsize = (width, height))
plt.plot(df['date'],df['capacityPillarOneTwo'], linestyle = 'solid', label = "Testing capacity")
plt.gcf().autofmt_xdate()
plt.grid()

plt.xlabel("Time")
plt.ylabel("Testing capacity")
plt.title("Pillar 1 and 2 testing capacity")
plt.legend()
plt.show()
```
